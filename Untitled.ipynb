# Analise de dados dos Candidatos do Estado de RJ 2020

```python
import pandas as pd #importando a biblioteca pandas para analisar os dados
import matplotlib.pyplot as plt #importando a biblioteca matplotlib para fazer os gráficos
```

```python
vEleicoes = pd.read_csv("C:/Users/João-Vitor/analise-dados/dados/consulta_cand_2020_RJ.csv", sep=";", encoding='Latin-1') #carregando o arquivo de dados dos Candidatos do estado de SP
```

```python
vEleicoes.head() #Ele lista as primeiras 4 linhas e colunas
```

```python
vEleicoes #Lista o total de linhas que temos no arquivo
```

```python
vEleicoes.columns.values #Listando as colunas que queremos
```

```python
ColunasSelecionada= ['NM_UE','NM_URNA_CANDIDATO',
       'SG_PARTIDO', 'DS_GENERO','DS_GRAU_INSTRUCAO'] #Selecionamos apenas as colunas que queriamos
```

```python
vEleicoesSelecionado = vEleicoes.filter(items=ColunasSelecionada) #Armazenar essa nova tabela
```

```python
vEleicoesSelecionado.head() #Nossa tabela filtrada ficou assim
```

```python
colunaPartido = vEleicoesSelecionado['SG_PARTIDO'] #Vamoslsitar colunas especifica de partidos
```

```python
colunaPartido #Aqui listamos até o final a quantidade de partidos do RJ
```

```python
colunaPartido.value_counts().sort_index() #Aqui agrupamos eles em grupos para saber a quantidade maxima pra cara variação
```

```python
#Agora que temos uma noção dos dados que analisamos vamos montar eles em graficos
```

```python
plt.figure(figsize = (50,20))
vEleicoes["SG_PARTIDO"].hist(bins = 30, ec = "k", alpha = .5, color = "Orange")
plt.suptitle('Partido dos Candidatos do Estado de RJ',  
             ha = 'center', 
             fontsize = 50, 
             weight = 'extra bold')

plt.xlabel("Partidos")
plt.ylabel("Quantidade de Candidatos")
```
